```python
import json
import os
import random
import string
import hashlib
import getpass

# File where account details will be stored
FILE_NAME = "password_vault.json"

# Master password for opening the vault
MASTER_PASSWORD = "admin123"


# -------------------------------
# Load saved accounts
# -------------------------------
def load_accounts():
    if os.path.exists(FILE_NAME):
        try:
            with open(FILE_NAME, "r") as file:
                return json.load(file)
        except:
            return {}
    return {}


# -------------------------------
# Save accounts
# -------------------------------
def save_accounts(accounts):
    with open(FILE_NAME, "w") as file:
        json.dump(accounts, file, indent=4)


accounts = load_accounts()


# -------------------------------
# Generate strong password
# -------------------------------
def generate_password():
    print("\n--- Strong Password Generator ---")

    try:
        length = int(input("Enter password length (minimum 8): "))
    except ValueError:
        print("❌ Please enter a number.")
        return

    if length < 8:
        print("❌ Password must contain at least 8 characters.")
        return

    characters = (
        string.ascii_uppercase
        + string.ascii_lowercase
        + string.digits
        + "!@#$%^&*"
    )

    password = "".join(random.choice(characters) for _ in range(length))

    print("\n🔐 Generated Password:", password)


# -------------------------------
# Add new account
# -------------------------------
def add_account():
    print("\n--- Add New Account ---")

    website = input("Website/App name: ").strip().lower()
    username = input("Username/Email: ").strip()
    password = getpass.getpass("Password: ")

    if website == "" or username == "" or password == "":
        print("❌ Fields cannot be empty.")
        return

    accounts[website] = {
        "username": username,
        "password": password
    }

    save_accounts(accounts)

    print("✅ Account saved successfully!")


# -------------------------------
# Search account
# -------------------------------
def search_account():
    website = input("\nEnter website/app name to search: ").strip().lower()

    if website in accounts:
        print("\n🔎 Account Found!")
        print("Website :", website.title())
        print("Username:", accounts[website]["username"])

        show = input("Show password? (yes/no): ").lower()

        if show == "yes":
            print("Password:", accounts[website]["password"])
        else:
            print("Password: ********")

    else:
        print("❌ Account not found.")


# -------------------------------
# View all accounts
# -------------------------------
def view_accounts():
    print("\n--- Saved Accounts ---")

    if not accounts:
        print("No accounts saved yet.")
        return

    for number, (website, details) in enumerate(accounts.items(), start=1):

        print(f"\n{number}. 🌐 {website.title()}")
        print("   Username:", details["username"])
        print("   Password: ********")


# -------------------------------
# Update account
# -------------------------------
def update_account():
    website = input("\nEnter website/app to update: ").strip().lower()

    if website not in accounts:
        print("❌ Account not found.")
        return

    print("Leave username blank to keep the old username.")

    new_username = input("New username/email: ").strip()
    new_password = getpass.getpass(
        "New password (leave blank to keep old password): "
    )

    if new_username:
        accounts[website]["username"] = new_username

    if new_password:
        accounts[website]["password"] = new_password

    save_accounts(accounts)

    print("✅ Account updated successfully!")


# -------------------------------
# Delete account
# -------------------------------
def delete_account():
    website = input("\nEnter website/app to delete: ").strip().lower()

    if website not in accounts:
        print("❌ Account not found.")
        return

    confirm = input(
        f"Are you sure you want to delete {website.title()}? (yes/no): "
    ).lower()

    if confirm == "yes":
        del accounts[website]
        save_accounts(accounts)
        print("🗑️ Account deleted successfully!")
    else:
        print("Deletion cancelled.")


# -------------------------------
# Vault statistics
# -------------------------------
def vault_statistics():
    print("\n--- Vault Statistics ---")
    print("📊 Total saved accounts:", len(accounts))

    if accounts:
        print("🌐 Saved websites/apps:")

        for website in accounts:
            print(" •", website.title())


# -------------------------------
# Master password login
# -------------------------------
def login():
    print("=" * 45)
    print("       🔐 SECURE PASSWORD VAULT")
    print("=" * 45)

    attempts = 3

    while attempts > 0:

        entered_password = getpass.getpass(
            "Enter Master Password: "
        )

        if entered_password == MASTER_PASSWORD:
            print("\n✅ Access Granted!")
            return True

        attempts -= 1
        print(f"❌ Wrong password! Attempts left: {attempts}")

    print("\n🚫 Too many failed attempts.")
    return False


# -------------------------------
# Main menu
# -------------------------------
def main():

    if not login():
        return

    while True:

        print("\n" + "=" * 45)
        print("       🔐 SECURE PASSWORD VAULT")
        print("=" * 45)

        print("1. ➕ Add New Account")
        print("2. 🔎 Search Account")
        print("3. 📋 View All Accounts")
        print("4. 🎲 Generate Strong Password")
        print("5. ✏️ Update Account")
        print("6. 🗑️ Delete Account")
        print("7. 📊 Vault Statistics")
        print("8. 🚪 Exit")

        choice = input("\nChoose an option (1-8): ")

        if choice == "1":
            add_account()

        elif choice == "2":
            search_account()

        elif choice == "3":
            view_accounts()

        elif choice == "4":
            generate_password()

        elif choice == "5":
            update_account()

        elif choice == "6":
            delete_account()

        elif choice == "7":
            vault_statistics()

        elif choice == "8":
            print("\n🔒 Vault locked successfully!")
            print("Thank you for using Secure Password Vault.")
            break

        else:
            print("❌ Invalid option. Please choose 1-8.")


# Start the program
main()
```
